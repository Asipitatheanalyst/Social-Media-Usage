```python
#visualization
#Daily Minutes Spent Per App
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Distribution of daily minutes spent  per App
plt.figure(figsize=(10, 6))
sns.boxplot(x='App', y='Daily_Minutes_Spent', data=df)
plt.title('Distribution of Daily Minutes Spent per App')
plt.xticks(rotation=45)
plt.show()
```

```python
#correlation between likes per day and post per day

plt.figure(figsize=(10, 5))
sns.barplot(x='App', y='Likes_Per_Day', data=df, estimator='mean', errorbar=None)
plt.title('Average Likes Per Day by App')
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(x='App', y='Posts_Per_Day', data=df, estimator='mean', errorbar=None)
plt.title('Average Posts Per Day by App')
plt.show()
```

```python
#Top Apps Based On User Activity(Likes

Likes_Per_Day = df.groupby('App')['Likes_Per_Day'].sum()

# Plotting a pie chart
plt.figure(figsize=(10, 5))
plt.pie(Likes_Per_Day , labels=Likes_Per_Day .index, autopct='%1.1f%%', startangle=140, colors=['yellowgreen', 'gold', 'lightcoral', 'lightskyblue', 'lightpink', 'lightblue'])
plt.axis('equal')
plt.title('Top Apps Based on User Activity')
plt.show()
```
